# The Symbolic Brain: Inside HNS-DS

This notebook explores the **Mental Model** of the Hybrid Neuro-Symbolic Deliberative Solver. Unlike black-box Large Language Models (LLMs), HNS-DS uses a transparent, symbolic "brain" to solve problems through formal reasoning and mathematical proof.

## 1. What is the "Mind" of HNS-DS?
In HNS-DS, the "mind" is not a set of weights in a neural network. Instead, it is a structured internal state called the **Mental Model**. It consists of:
- **Logic Mind (Formalizer)**: Translates human language into math.
- **Executive Mind (Planner)**: Strategizes how to break down the task.
- **Reasoning Mind (Synthesizer & Verifier)**: Hypothesizes solutions and formally proves them using Z3.
- **Memory (Learner)**: Recalls past successes and failures.

```python
import logging
import sys
from hnsds.orchestrator import HNSDSOrchestrator
from hnsds.perception.parser import Parser
from hnsds.formalizer.spec_builder import SpecBuilder
from hnsds.planner.htn_planner import HTNPlanner
from hnsds.synthesizer.enumerative import EnumerativeSynthesizer
from hnsds.verifier.z3_interface import Z3Verifier
from hnsds.learner.episode_logger import EpisodeLogger

# 1. Initialize the system components
perception = Parser()
formalizer = SpecBuilder()
planner = HTNPlanner()
verifier = Z3Verifier()
learner = EpisodeLogger()
synthesizer = EnumerativeSynthesizer(learner=learner, search_range=(-10, 10))

# 2. The Orchestrator is the manager of the RIR-RI loop
orchestrator = HNSDSOrchestrator(perception, formalizer, planner, synthesizer, verifier, learner)
```

## 2. Solving a Complex Problem
Let's see the AI solve a system of equations. We'll watch its "mind" in action as it proposes candidates and verifies them.

```python
# Define a problem
problem_input = "Solve x + y = 15, x - y = 3"

print(f"User Input: {problem_input}")
solution = orchestrator.run(problem_input)

if solution:
    print(f"\nProven Solution: {solution}")
else:
    print("\nStatus: Failed to find a verified proof.")
```

## 3. Visualizing the Mental Model
The `Deliberation Trace` is the literal output of the AI's internal reasoning state. It shows exactly how it "thought" about the problem.

```python
from IPython.display import Markdown
Markdown(orchestrator.write_solution())
```

## 4. Why this is a New Paradigm
This AI doesn't "guess" tokens. If you ask the same problem again, it uses its **Learner** to retrieve the past success instantly, just like human intuition, but backed by a formal proof.

Every step is **Deterministic, Sound, and Traceable.**
